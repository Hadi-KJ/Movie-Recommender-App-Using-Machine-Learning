```python
import ast #abstract syntax tree
```

```python
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
movies_df['keywords'] = movies_df['keywords'].apply(convert)
movies_df.head()
```

```python
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter !=3:
      L.append(i['name'])
      counter +=1
    else:
      break
  return L
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert3)
movies_df.head()
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies_df['crew'] = movies_df['crew'].apply(fetch_director)
movies_df.head()
```

```python
movies_df['overview'] = movies_df["overview"].apply(lambda x:x.split())
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies_df['cast'] = movies_df['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies_df['crew'] = movies_df['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies_df
```

```python
movies_df['tags'] = movies_df['overview'] + movies_df['genres'] + movies_df['keywords']+ movies_df['cast'] + movies_df['crew']
```

```python
movies_df
```

```python
new_df = movies_df[['movie_id', 'title', 'tags']]
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#the library to transform a given text into a vector on the basis of the frequency count of each word in the entire text
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
#natural language tool kit
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ". join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#similarity between two vectors of an inner product space, it is measured by the angles between two vectors
# to measure document similarity in text analysis
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sum(similarity[0])
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])), reverse= True, key=lambda x:x[1])[1:6]
```

```python
print(list(enumerate(similarity[0]))[1:6])
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse= True, key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Iron Man')
```
